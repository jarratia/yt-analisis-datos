```python
import pandas as pd
import numpy as np
```

```python
#Data frame
data = {
  "Nombre":["Ana", "Juan", "Pedro", "María", "Luis", "Pepe"],
  "Edad":[22, 25, 28, 23, 20, 25],
  "Ciudad":["Barcelona", "Caracas", "Madrid", "Santiago de Chile", "Buenos Aires", "Caracas"],
  "Puntuación": [80, 90, 85, 100, 75, 91]
}
df = pd.DataFrame(data)
df
```

```python
#Supongamos que deseamos agrupar el grupo de personas para conocer los habitantes por ciudad
grouped = df.groupby("Ciudad")

print(grouped.groups)
```

```python
aggregated_data = grouped.agg(
  {
    "Edad": "mean",
    "Puntuación":"sum"
  }
)
aggregated_data
```

```python
#Agregando custom functions
#Agregación personalizada donde calculemos el rango de edades

def rango(series):
  return series.max() - series.min()

#Aplicamos la función al grupo
aggregated_data_custom = grouped.agg(
  {
    "Edad":rango,
    "Puntuación": rango
  }
)

aggregated_data_custom
```

```python
#Agrupación en múltiples niveles
data["Categoria"] = ["A", "B", "A", "B", "A", "B"]
df = pd.DataFrame(data)
df
```

```python
#Agrupar datos por ciudad y categoría (Multinivel)
grouped_multi = df.groupby(["Ciudad", "Categoria"])
grouped_multi.groups
```

```python
#Agregación de calculo de suma de edades y puntuaciones por ciudad y por categoria
aggregated_data_multi = grouped_multi.agg(
  {
    "Edad":"sum",
    "Puntuación":"mean"
  }
)
aggregated_data_multi
```
